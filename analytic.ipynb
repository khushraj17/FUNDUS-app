```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
latlong = pd.read_csv('latlong.csv')
```

```python
latlong.head()
```

```python

# Function to extract latitude and longitude
def extract_lat_lon(coord):
    try:
        lat, lon = coord.split(",")

        lat = float(lat.replace("° N", "").replace("° S", "").strip())
        lon = float(lon.replace("° E", "").replace("° W", "").strip())

        # South aur West ko negative bana do
        if "S" in coord:
            lat = -lat
        if "W" in coord:
            lon = -lon

        return pd.Series([lat, lon])

    except:
        return pd.Series([None, None])
```

```python

latlong[["latitude", "longitude"]] = latlong["coordinates"].apply(extract_lat_lon)
```

```python
latlong.drop(columns=["coordinates"], inplace=True)
```
